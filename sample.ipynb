```python
import requests
import pandas as pd

test_data = pd.read_csv('data/CMAPSSData/test_FD001.txt', sep=r'\s+', header=None,
    names=['unit','cycle','op_setting_1','op_setting_2','op_setting_3'] + [f'sensor_{i}' for i in range(1,22)])

engine_1_last5 = test_data[test_data['unit'] == 1].tail(5)
sensor_cols = [f'sensor_{i}' for i in range(1, 22)]
payload = {"recent_cycles": engine_1_last5[sensor_cols].to_dict('records')}

response = requests.post('http://localhost:8000/predict', json=payload)
print(response.status_code)
print(response.json())
```
